## Fake news detection

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import re
import string
```

```python
df_fake = pd.read_csv("Fake.csv")
df_true = pd.read_csv("True.csv")
```

```python
df_fake.head(10)
```

```python
df_true.head(10)
```

```python
df_fake['class'] = 0
df_true['class'] = 1
```

```python
df_fake.shape
```

```python
df_true.shape
```

```python
df_fake_manual_testing = df_fake.tail(10)
for i in range(23480,23470, -1):
    df_fake.drop([i],axis=0,inplace=True)
df_true_manual_testing = df_true.tail(10)
for i in range(21416, 21406, -1):
    df_true.drop([i],axis=0,inplace=True)    
```

```python
df_manual_testing = pd.concat([df_fake_manual_testing,df_true_manual_testing], axis=0)
df_manual_testing.to_csv("manual_testing.csv")
```

```python
df_merge = pd.concat([df_fake,df_true],axis=0)
df_merge.head()
```

```python
df = df_merge.drop(["title","subject","date"],axis=1)
df.head(10)
```

```python
df = df.sample(frac=1)
df.head()
```

```python
df.isnull().sum()
```

```python
def word_drop(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub("\\ W"," ",text)
    text = re.sub('https?://\S+/www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub(' [%s]' % re.escape(string.punctuation),'' ,text)
    text = re.sub('\n ', '', text)
    text = re.sub('\w*\d\w*','', text)
    return text
```

```python
df['text'] = df['text'].apply(word_drop)
```

```python
df.head()
df
df.value_counts()
```

```python
x = df['text']
y = df['class']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.feature_extraction.text import HashingVectorizer

hashing_vectorizer = HashingVectorizer()
x_train_hashed = hashing_vectorizer.transform(x_train)
x_test_hashed = hashing_vectorizer.transform(x_test)
```

```python
x_train,x_test
print("X_train shape:", x_train.shape)
print("X_test shape:", x_test.shape)

# Additional debugging information
print("Y_train shape:", y_train.shape)
print("Y_test shape:", y_test.shape)
```

## Since we have done with preprocessing and spliting our data now we will build classification models for our classification problem

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
Gbc = GradientBoostingClassifier(random_state=0)
Gbc.fit(x_train_hashed,y_train)
```

```python
Gbc.predict(x_test_hashed)
```

```python
Gbc.score(x_test_hashed,y_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train_hashed,y_train)
```

```python
lr.score(x_test_hashed,y_test)
```

```python
def output_lable(n):
    if n == 0:
        return "Fake News"
    elif n == 1:
        return "Not A Fake News"


def manual_testing(news):
    testing_news = {"text":[news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(word_drop)
    new_x_test = new_def_test["text"]
    new_xhashed_test = hashing_vectorizer.transform(new_x_test)
    pred_lr = lr.predict(new_xhashed_test)
    pred_gbc = Gbc.predict(new_xhashed_test) 
    
    return print("\n\nLR Prediction: {} \nGBC Prediction: {}".format(output_lable(pred_lr[0]),output_lable(pred_gbc[0])))






```
